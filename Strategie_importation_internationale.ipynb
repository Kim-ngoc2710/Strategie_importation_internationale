## **II. Explication de la choix des variables**


**- Population totale** : Mesurer la taille du marché potentiel de chaque pays. Plus la population est importante, plus la demande potentielle peut être élevée.


**- Taux d’urbanisation** : Comprendre la structure de la population et les habitudes de consommation, souvent différentes entre zones urbaines et rurales.


**- PIB par habitant** : Evaluer le niveau de richesse moyen et le pouvoir d’achat des consommateurs.


**- Inflation** : cette variable donne une idée de la stabilité économique du pays. Une inflation élevée peut rendre le marché plus incertain.


**- Stabilité politique** : Prendre en compte le niveau de risque du pays dans une logique d’exportation.


**- Logistics Performance Index (LPI)** : Evaluer la qualité de la logistique et la facilité des échanges commerciaux.


**- Production de volaille** : Cette variable permet de voir si le pays produit déjà beaucoup de volaille localement.


**- Importations de volaille** : Cette variable aide à repérer les pays qui dépendent déjà des importations pour répondre à leur demande.


**- Exportations de volaille** : Identifier les pays qui sont déjà compétitifs dans ce secteur.


**- Disponibilité intérieure de volaille** : Cette variable donne une vision globale de la quantité de volaille disponible sur le marché intérieur.

## **VII. Principal Component Analysis (PCA)**

### **1. PCA**

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

**L’ACP permet de créer jusqu’à 10 composantes principales (Dim1 à Dim10), chacune étant une combinaison linéaire des variables initiales.**

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=[f"Dim{i+1}" for i in range(len(Variable))],
    index=Variable
)

loadings
```

### **2. Choisir le nombre de dimensions**

```python
explained_var = pca.explained_variance_ratio_
```

```python
cum_var = pca.explained_variance_ratio_.cumsum()
```

```python
plt.figure(figsize=(15,10))

# variance chacun dimension
plt.plot(range(1, len(explained_var)+1), explained_var, marker='o', label="Variance expliquée")

# cumulative variance
plt.plot(range(1, len(cum_var)+1), cum_var, marker='s', linestyle='--', label="Variance cumulée")

plt.xlabel("Dim")
plt.ylabel("Pourcentage d'info présentantes")
plt.title("Éboulis des valeurs propres")
plt.legend()
plt.grid(True)

# Enlever les bordures
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

**Les deux premières composantes principales expliquent environ 70 % de la variance totale, ce qui permet déjà une bonne représentation des données. L’ajout de la troisième composante augmente cette proportion à près de 80 %, apportant ainsi un gain d’information non négligeable. Le choix du nombre de dimensions dépend donc du compromis entre simplicité de visualisation et richesse de l’information conservée.**

**Pour représenter les pays sur le cercle des corrélations et le plan factoriel, nous retenons deux dimensions afin de faciliter la visualisation, tout en conservant environ 70 % de l’information initiale.**

```python
loadings[['Dim1','Dim2']]
```

```python
plt.figure(figsize=(10,10))

for i, var in enumerate(Variable):
    x = loadings.loc[var, 'Dim1']
    y = loadings.loc[var, 'Dim2']

    plt.arrow(0, 0, x, y, head_width=0.03)
    plt.text(x*1.1, y*1.1, var, fontsize=9)

# vẽ cercle
circle = plt.Circle((0,0), 1, color='blue', fill=False)
plt.gca().add_artist(circle)

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.xlim(-1,1)
plt.ylim(-1,1)

plt.title("Cercle des corrélations")
plt.xlabel("Dim1")
plt.ylabel("Dim2")

plt.grid(False)

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(False)

plt.show()
```

**Le cercle des corrélations permet d’interpréter les axes factoriels. Le premier axe est principalement associé aux variables économiques et logistiques, tandis que le second oppose la taille du marché au niveau de développement.**

**Le premier axe (Dim1) semble traduire un niveau global de développement économique et d’intégration logistique. Il est fortement corrélé positivement avec des variables telles que le PIB par habitant, les exportations, la production et l’indice logistique (LPI), et négativement avec l’inflation (-0.11). Ainsi, les pays situés du côté positif de cet axe correspondent plutôt à des économies développées, performantes sur le plan logistique et caractérisées par une inflation relativement faible.**

**Le deuxième axe (Dim2) oppose des pays caractérisés par une grande taille de marché (population élevée, production et disponibilité intérieure importantes) à des pays davantage développés sur le plan économique et institutionnel (PIB par habitant élevé, stabilité politique, meilleure performance logistique). Cet axe met donc en évidence une opposition entre la dimension quantitative (taille et volume) et la dimension qualitative (niveau de développement).**

```python
X_pca_df = pd.DataFrame(
    X_pca[:, :2],
    columns=['Dim1', 'Dim2']
)

X_pca_df['Country'] = fact_2017['Zone_FR'].values
```

```python
X_pca_df
```

```python
def quadrant(row):
    if row['Dim1'] >= 0 and row['Dim2'] >= 0:
        return 'Q1'
    elif row['Dim1'] < 0 and row['Dim2'] >= 0:
        return 'Q2'
    elif row['Dim1'] < 0 and row['Dim2'] < 0:
        return 'Q3'
    else:
        return 'Q4'

X_pca_df['Quadrant'] = X_pca_df.apply(quadrant, axis=1)
```

```python
label_map = {
    'Q1': 'Forts & grands marchés',
    'Q2': 'Grands marchés émergents',
    'Q3': 'Petits marchés peu performants',
    'Q4': 'Marchés développés (taille modérée)'
}

X_pca_df['Segment'] = X_pca_df['Quadrant'].map(label_map)
```

## **VIII. Classification ascendante hiérachique (CAH)**

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
```

**Pour l’analyse de clustering et CAH, nous retenons les trois premières composantes principales. En effet, celles-ci permettent de capturer une part plus importante de la variance totale (environ 80 %), ce qui améliore la fiabilité de la représentation des données. L’ajout de la troisième dimension permet ainsi de préserver davantage d’information tout en réduisant le bruit et la redondance des variables initiales.**

```python
X_clust = X_pca[:, :3]
```

```python
#La méthode de Ward regroupe les observations en minimisant la variance au sein des clusters, ce qui permet d’obtenir des groupes homogènes.

Z = linkage(X_clust, method='ward')
```

```python
plt.figure(figsize=(15, 10))
dendrogram(Z)
plt.title("Dendrogramme - CAH")
plt.xlabel("Pays")
plt.ylabel("Distance")
plt.show()
```

**Le dendrogramme met en évidence une structuration des données en environ 3 à 4 groupes distincts.**

```python

clusters_cah_3 = fcluster(Z, 3, criterion='maxclust')
clusters_cah_4 = fcluster(Z, 4, criterion='maxclust')


X_pca_df['cluster_CAH_3'] = clusters_cah_3.astype(str)
X_pca_df['cluster_CAH_4'] = clusters_cah_4.astype(str)
```

```python
from sklearn.metrics import silhouette_score

score_3 = silhouette_score(X_clust, clusters_cah_3)
score_4 = silhouette_score(X_clust, clusters_cah_4)

print("Silhouette CAH - 3 clusters :", round(score_3, 3))
print("Silhouette CAH - 4 clusters :", round(score_4, 3))
```

**Le coefficient de silhouette a été utilisé pour évaluer la qualité de la segmentation. L’évaluation de la qualité de la segmentation à l’aide du coefficient de silhouette montre des résultats proches entre les solutions à 3 et 4 clusters (0.307 contre 0.316).**

```python
from scipy.cluster.hierarchy import fcluster
import seaborn as sns




fig, axes = plt.subplots(1, 2, figsize=(20, 7), sharex=True, sharey=True)

sns.scatterplot(
    data=X_pca_df,
    x='Dim1',
    y='Dim2',
    hue='cluster_CAH_3',
    hue_order=['1', '2', '3'],
    palette='tab10',
    s=70,
    ax=axes[0]
)
axes[0].axhline(0, color='black', linewidth=1)
axes[0].axvline(0, color='black', linewidth=1)
axes[0].set_title("CAH - 3 clusters")
axes[0].set_xlabel("Axe 1 : économie / logistique")
axes[0].set_ylabel("Axe 2 : taille / développement")
axes[0].grid(False)

sns.scatterplot(
    data=X_pca_df,
    x='Dim1',
    y='Dim2',
    hue='cluster_CAH_4',
    hue_order=['1', '2', '3', '4'],
    palette='tab10',
    s=70,
    ax=axes[1]
)
axes[1].axhline(0, color='black', linewidth=1)
axes[1].axvline(0, color='black', linewidth=1)
axes[1].set_title("CAH - 4 clusters")
axes[1].set_xlabel("Axe 1 : économie / logistique")
axes[1].set_ylabel("")
axes[1].grid(False)


for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)


plt.subplots_adjust(wspace=0.25, right=0.92)
plt.tight_layout()
plt.show()
```

**La partition en 3 clusters apparaît plus lisible et plus cohérente que celle en 4 groupes. En effet, avec k=3, les groupes obtenus sont mieux séparés sur le plan factoriel et présentent une structure plus claire, tandis que la solution en 4 clusters introduit une fragmentation supplémentaire sans réel gain d’interprétation. Dans une logique de segmentation des pays, la solution à 3 groupes offre donc un meilleur compromis entre lisibilité, stabilité et pertinence analytique.**
